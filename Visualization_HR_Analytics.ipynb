```python
#Let's create a new feature that is a Amount with log to better see the values distribuitions
tr['Score_log'] = np.log(tr['Score'] + 1)

plt.figure(figsize=(8,5))
sns.distplot(tr['Score_log'].dropna())
plt.xlabel('log of Score', fontsize=12)
plt.title("Log Hist of Score", fontsize=16)
plt.show()
```

```python
#Let's create a new feature that is a Amount with log to better see the values distribuitions
tr['Work Experience_log'] = np.log(tr['Work Experience'] + 1)

plt.figure(figsize=(8,5))
sns.distplot(tr['Work Experience_log'].dropna())
plt.xlabel('log of Work Experience', fontsize=12)
plt.title("Log Hist of Work Experience", fontsize=16)
plt.show()
```

```python
#Let's create a new feature that is a Amount with log to better see the values distribuitions
tr['CGPA (Ist & IInd Year)_log'] = np.log(tr['CGPA (Ist & IInd Year)'] + 1)

plt.figure(figsize=(8,5))
sns.distplot(tr['CGPA (Ist & IInd Year)_log'].dropna())
plt.xlabel('log of CGPA (Ist & IInd Year)', fontsize=12)
plt.title("Log Hist of CGPA (Ist & IInd Year)", fontsize=16)
plt.show()
```

```python
#Let's create a new feature that is a Amount with log to better see the values distribuitions
tr['Age_log'] = np.log(tr['Age'] + 1)

plt.figure(figsize=(8,5))
sns.distplot(tr['Age_log'].dropna())
plt.xlabel('log of Age', fontsize=12)
plt.title("Log Hist of Age", fontsize=16)
plt.show()
```
